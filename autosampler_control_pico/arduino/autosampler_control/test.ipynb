```python
# Full tanh curve-based interpolation for ramp profile

import numpy as np
import matplotlib.pyplot as plt

# Parameters
RAMP_SIZE = 200
MAX_INTERVAL = 3000
MIN_INTERVAL = 500
SCALE = 3.0  # Controls how wide the tanh curve is (larger = steeper transition)

# Full tanh: x in [-scale, +scale]
x = np.linspace(0, SCALE, RAMP_SIZE)
tanh_vals = np.tanh(x)  # now spans from -1 to +1
normalized = (tanh_vals + 1) / 2  # map to [0, 1]
# invert so that left is MAX, right is MIN
ramp_profile = MIN_INTERVAL + (1 - tanh_vals) * (MAX_INTERVAL - MIN_INTERVAL)
# keep only the integers
ramp_profile = np.round(ramp_profile).astype(int)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(RAMP_SIZE), ramp_profile, marker="o", label="Tanh Ramp Profile")
plt.title("Full Tanh-Based Ramp Profile")
plt.xlabel("Step Index")
plt.ylabel("Time Interval (e.g., Î¼s)")
plt.grid(True)
plt.legend()
plt.show()
```
